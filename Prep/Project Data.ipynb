```python
import pandas as pd
```

```python
dat = pd.read_csv('used_cars_data.csv')
dat.head()
```

```python
na_counts = pd.DataFrame(dat.isna().sum())
```

```python
na_counts = na_counts.reset_index()
na_counts
```

```python
na_counts['cols'] = na_counts['index']
```

```python
cols_remove = (na_counts.sort_values(by = 0, ascending = False)).head(9)
```

```python
#removed columns with 2 million missing
for i in (list(cols_remove.cols)):
    dat = dat.drop(i, axis = 1)
```

```python
cols_remove.cols
```

```python
dat.isna().sum()
```

```python
nonmissing_dat = dat.dropna()
```

```python
nonmissing_dat.shape
```

```python
data = nonmissing_dat.sample(n = 6000, random_state = 10)
```
